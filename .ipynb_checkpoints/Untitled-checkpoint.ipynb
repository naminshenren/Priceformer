```python
data = pd.read_csv(r"./dataset/ETT-small/ETTh2.csv")
data = data.rename(columns={'date':'ds','OT':'y'}) 
max_min_scaler = lambda x : (x-np.mean(x))/(np.std(x))
data[['y']]=data[['y']].apply(max_min_scaler)
```

```python
def get_pre(index,seqlen,prelen,index2):
    data_sample = data.iloc[0+index+index2:index+seqlen+index2]
    #future = train.iloc[index+seqlen:index+seqlen+prelen]
    
    #future.drop(future.columns[1:8], axis=1, inplace=True)
    model = prophet.Prophet()
    model.fit(data_sample)
    future = model.make_future_dataframe(periods=prelen,freq="H")
    future_results = model.predict(future)
    pre_vs_real = np.array(data.iloc[index+seqlen+index2:index+seqlen+prelen+index2]['y'])-np.array(future_results[-prelen:]['yhat'])
    return np.abs(pre_vs_real).sum(), np.power(pre_vs_real,2).sum(), prelen
```

```python
all_ae = 0.0
all_se = 0.0
all_num = 0
seqlen = 336
prelen = 720
index2 = 12 * 30 * 24 + 4 * 30 * 24 - seqlen
#train = data.iloc[12 * 30 * 24 * 4 + 4 * 30 * 24 * 4 - seqlen:12 * 30 * 24 * 4 + 4 * 30 * 24 * 4]
for i in range(seqlen):
    ae, se, number = get_pre(i, seqlen, prelen, index2)
    all_ae += ae
    all_se += se
    all_num += number
print("mse: ", all_se/all_num, " mae: ", all_ae/all_num)
```
